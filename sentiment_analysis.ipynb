```python
import pandas as pd

from germansentiment import SentimentModel
```

```python
df = pd.read_csv('/content/cnn_predicted_trans_train-2.csv', sep = ',')
```

```python
model = SentimentModel()
```

```python
import pandas as pd
import numpy as np

# Function for sentiment mapping
def map_sentiment(sentiment):
    sentiment_map = {'negative': -1, 'positive': 1, 'neutral': 0}
    return sentiment_map.get(sentiment, None)



# Split comments and predict sentiment
df['Comment'] = df['Comment'].astype(str)
list_text = [sentence.split('.') for comment in df['Comment'] for sentence in comment.split('\n')]
result_list = [[map_sentiment(model.predict_sentiment(sentence)) for sentence in comment] for comment in list_text]

# Function for sentiment calculation
def calculate_sentiment(sentiment_list):
    return [np.mean(sentiments) for sentiments in sentiment_list]

# Calculate sentiment and assign to DataFrame
df['Sentiment'] = calculate_sentiment(result_list)

# Sample the DataFrame
df_sampled = df.sample(n=999, replace=True, random_state=1)

# Save to CSV
df_sampled.to_csv('cnn_sentiment_trans_train.csv', index=False)
```
